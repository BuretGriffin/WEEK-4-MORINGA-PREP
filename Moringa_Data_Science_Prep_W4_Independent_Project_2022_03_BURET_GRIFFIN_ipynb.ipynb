```python
# importing  numpy, pandas and time
import pandas as pd

import numpy as np

from datetime import datetime
```

```python
# data cleaning
# dropping uncessary columns
df3.drop(['Displayed comment', 'ID', 'Geo point','Address', 'Cars'], axis = 1, inplace = True)
df3
```

```python
# checking and dropping duplicates
df3[df3.duplicated()]

# dropping the duplicates

df =df3.drop_duplicates()
 
# checking for non duplicates

df.head(20)
```

```python
# getting the  number of cars at the station

df['Bluecar_Diff'] = df['Bluecar counter'].diff()
df.head()
```

```python
# answering questions
# question1
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris
Electric_paris = df[df['City'] =='Paris'] 

Electric_paris.head()
```

```python


Electric_paris[Electric_paris['Bluecar_Diff'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```python
#What station is the most popular?
#Overall?
#At the most popular picking hour?
# Identifying the the most popular hour for returning cars
# Usage is positive when returning cars
#
Electric_paris[Electric_paris.Bluecar_Diff > 0].groupby('hour')['hour'].count().sort_values(ascending = False).head(1)
```
